```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import ipywidgets as widgets
from IPython.display import display, Markdown
from gtts import gTTS
from IPython.display import Audio
import os
def speak(text):
  tts = gTTS(text=text, lang='en')
  tts.save("output.mp3")
  display(Audio("output.mp3", autoplay=True))


project_name = Markdown('# Sentiment Analysis')
display(project_name)
def clearAll():
  global text
  text = ""
def detect_sentiment(sentence):
  sid_obj = SentimentIntensityAnalyzer()
  sentiment_dict = sid_obj.polarity_scores(sentence)
  negative = str(sentiment_dict['neg'] * 100) + "% Negative"
  neutral = str(sentiment_dict['neu'] * 100) + "% Neutral"
  positive = str(sentiment_dict['pos'] * 100) + "% Positive"
  if sentiment_dict['compound'] >= 0.05:
    overall = "Positive"
  elif sentiment_dict['compound'] <= -0.05:
    overall = "Negative"
  else:
    overall = "Neutral"
  return negative, neutral, positive, overall

def on_button_click(b):
  sentence = text_box.value
  negative, neutral, positive, overall = detect_sentiment(sentence)
  result = f"The sentence was rated as:\n{negative}\n{neutral}\n{positive}\nThe Sentence Overall Rated As: {overall}"
  output_text.value = result
  speak(result)


text_box = widgets.Textarea(
value='',
placeholder='Enter your text here',
description='Text:',
disabled=False,
layout=widgets.Layout(width='50%', height='80px')
)

output_text = widgets.Textarea(
value='',
placeholder='Sentiment analysis result will be displayed here',
description='Output:',
disabled=True,
layout=widgets.Layout(width='50%', height='100px')
)

button = widgets.Button(
description='Analyze Sentiment',
disabled=False,
button_style='',
tooltip='Click to analyze sentiment',
icon='check'
)

button.on_click(on_button_click)
display(text_box)
display(button)
display(output_text)

up_btn = Markdown(' Made by: Kanika Singh Deo')
display(up_btn)
```
